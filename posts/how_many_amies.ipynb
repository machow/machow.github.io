```python
import requests
import re
URL_TRANSCRIPTS = "https://www.springfieldspringfield.co.uk/episode_scripts.php?tv-show=criminal-minds"

r = requests.get(URL_TRANSCRIPTS)

body = r.text
```

```python
RE = r"""href="(view_episode_scripts\.php\?tv-show=criminal-minds&episode=)(s\d+e\d+)"""
matches = re.findall(RE, body, flags = re.MULTILINE)
```

```python
matches
# check that we didn't get duplicate episode links
print(len(set(x[1] for x in matches)))
print(len(matches))
```

```python
from bs4 import BeautifulSoup
base_url = "https://www.springfieldspringfield.co.uk/"
transcripts = []
for link, eps in matches:
    print(eps)
    transcript_page = requests.get(base_url + link + eps).text
    trans_text = BeautifulSoup(transcript_page).select(".scrolling-script-container")[0].text
    transcripts.append(trans_text)
    
```

```python
amy_counts = [any([x.count("Amy"), x.count("Aimee")]) for x in transcripts]

print("Amy episodes:", sum(amy_counts))
print("Total episodes:", len(amy_counts))
```

```python
import pandas as pd
import qgrid

df = pd.DataFrame({
    'meta': [x[1] for x in matches],
    'has_amy': amy_counts
    })

df['ep'] = df.meta.apply(lambda s: s[4:])
df['season'] = df.meta.apply(lambda s: s[1:3])
df.groupby('season').has_amy.sum()

df.query("has_amy")
#qgrid.show_grid(df)
```
